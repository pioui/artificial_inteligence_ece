Η γενική μορφή της συνάρτησης κόστους στον αλγόριθμο Α* είναι:
$$f(n)=g(n)+h(n)$$
Παραπάνω, η συνάρτηση $g(n)$ δίνει την πραγματική απόσταση από το σημείο εκκίνησης μέχρι τον κόμβο $n$, και η συνάρτηση $h(n)$ αποτελεί μια ευριστική της απόστασης από τον κόμβο $n$ μέχρι τον στόχο. Σας ζητείται να πειραματιστείτε με τις εξής επιλογές για τις δύο συναρτήσεις:
* $g(n)=0$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$. Ποιος αλγόριθμος αναζήτησης προκύπτει; Μπορεί να βρει πάντα το βέλτιστο μονοπάτι;
* $g(n)=1$ και $h(n)=0$. Ποιος αλγόριθμος αναζήτησης προκύπτει; Μπορεί να βρει πάντα το βέλτιστο μονοπάτι;
* $g(n)=1$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$. Ποιος αλγόριθμος αναζήτησης προκύπτει; Μπορεί να βρει πάντα το βέλτιστο μονοπάτι;

## Κώδικας visualization


```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος:
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []

    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος:
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[~grid.grid] = [0, 0, 0]
        image[grid.grid] = [255, 255, 255]
        # Use this to treat 1/True as obstacles
        # image[grid.grid] = [0, 0, 0]
        # image[~grid.grid] = [255, 255, 255]

        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)

    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)

    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani

    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
```

```
from queue import PriorityQueue

class pathfinder:
    def __init__(self, S, F, grid, c, h,visualize=True):
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        self.no_expanded_nodes = 0

        #for visualization
        self.visualize=visualize
        self.vis = visualization(S, F)
        self.find_path()


    def find_path(self):
        frontier = PriorityQueue()
        frontier.put((0, self.S))
        came_from = {}
        cost_so_far = {}
        came_from[self.S] = None
        cost_so_far[self.S] = 0
        self.no_expanded_nodes = 0

        while not frontier.empty():
            current = frontier.get()[1]
            if current == self.F:
                break
            self.no_expanded_nodes += 1

            for next in  self.grid.adjacent_no_walls(current):
                ### YOUR CODE HERE \/ (cost so far+cost to next)
                new_cost = cost_so_far[current] + 1
                ### YOUR CODE HERE /\

                if next not in cost_so_far or new_cost < cost_so_far[next]:
                    cost_so_far[next] = new_cost

                    ### YOUR CODE HERE \/ (c+h)
                    priority = self.cost(current,next)+self.heuristic(next,self.F)
                    ### YOUR CODE HERE /\

                    frontier.put((priority, next))
                    came_from[next] = current
                #for animation
                #find frontier and expanded nodes
                restore = []
                while not frontier.empty():
                    node = frontier.get()
                    restore.append(node)
                for node in restore:
                    frontier.put(node)
                exp = []
                #find expanded nodes
                for v in came_from:
                    if v:
                        exp.append(v)
                if self.visualize:
                  self.vis.draw_step(self.grid, [r[1] for r in restore], exp)


        # follow backlinks to build path
        # current = self.F
        path = []
        while current != self.S:
            path.append(current)
            if current not in came_from:
              self.grid.draw_map()
            current = came_from[current]
        path.append(self.S)
        path.reverse()

        self.path = path

    def get_path(self):
        return self.path

    def get_complexity(self):
        return self.no_expanded_nodes
```

### AStar

 AStar algorithm also maintains a priority queue of the nodes to be expanded next, but in addition to the distance from the start node, it uses a heuristic function to estimate the distance from each node to the goal. The heuristic function is an admissible heuristic, meaning that it never overestimates the actual distance to the goal. This allows AStar to prioritize nodes that are more likely to lead to the goal, rather than exploring all possible paths as Dijkstra's algorithm does. \\

However, there are certain situations where AStar algorithm may not always find the shortest path. One such situation is when the heuristic function is not admissible, and it overestimates the actual distance to the goal. In such cases, AStar may instead return a suboptimal path.

#### Admissible heuristics
An admissible heuristic function is a function used in AStar algorithm that never overestimates the actual distance to the goal node. In other words, the heuristic function should always estimate the shortest possible distance to the goal, or at most, the same distance as the actual shortest path. There are several types of admissible heuristic functions that can be used in AStar algorithm, depending on the specific problem and the information available. Here are some examples:

- Euclidean distance heuristic: This heuristic calculates the straight-line distance between two points, which is the shortest possible distance between them. This heuristic is admissible as it always underestimates or equals the actual distance.

- Manhattan distance heuristic: This heuristic is commonly used in grid-based search problems, such as pathfinding on a map. It calculates the distance between two points by summing the absolute differences in their x and y coordinates. This heuristic is admissible as it never overestimates the actual distance.

- Chebyshev distance heuristic calculates the distance between two points in a grid-like environment by taking the maximum of the absolute differences in their x and y coordinates. This heuristic is also known as the "chessboard" distance or the "king's move" distance, as it mimics the movement of a king in chess.
- Diagonal distance heuristic: This heuristic is similar to the Manhattan distance heuristic, but takes into account diagonal movement. It calculates the distance by taking the maximum of the absolute differences in the x and y coordinates. This heuristic is admissible as it never overestimates the actual distance. It's not well applicable in our case though, since there is no diagonal movement in the maze

### Dijkstra

By setting heuristic to 0 we only consider distance from the parent node. Dijkstra's algorithm explores all possible paths from the start node to the goal node, maintaining a priority queue of the nodes to be expanded next based on their distances from the start node. It visits each node only once, and when a node is visited, it updates the distances of its neighbors. Dijkstra's algorithm guarantees that the shortest path is found once the goal node is reached.

### Best-first search

Setting the cost of moving from the starting node to the current node g to 0 for all nodes in the AStar algorithm will cause the algorithm to behave like a greedy best-first search, where it only considers the heuristic estimates of the cost of moving from the current node to the goal node.

In other words, the algorithm will prioritize exploring nodes that are closer to the goal node based on the heuristic function, without taking into account the actual cost of moving from the starting node to the current node. This can lead to suboptimal paths being returned by the algorithm, as it may overlook longer paths that have a lower total cost than shorter paths.

We need to come up with a heuristic function that estimates the value of a given game state, i.e., the score of each move. The intuition behind this heuristic is that the agent wants to move towards the goal while staying away from the ghost. This means that the score needs to satisfy the following two conditions:

- The closer the agent is to the target, the higher the score should be (and vice-versa).
- The farther away the agent is from the ghost, the higher the score should be (and vice-versa).

To construct this heuristic, we first need to define the distances. In general, Manhattan distance is more appropriate when the agent can only move in horizontal or vertical directions, while Euclidean distance is more appropriate when the agent can move diagonally or in any direction. Since the maze is constructed in such a way that diagonal movement is not possible, Manhattan distance is a better choice.

We define $TD$ as the Manhattan distance from the target and $GD$ as the Manhattan distance from the ghost. Additionally, we set the density of the maze to $ 0.2$. To avoid oscillations in the same spots in `adjacent_no_walls`, we shuffle the possible next steps.

We conducted several experiments to evaluate different heuristic functions.

```
from IPython.display import HTML, Image
from google.colab.output import eval_js
from base64 import b64decode
import PIL


canvas_html = """
<canvas width=301 height=301></canvas>
<br>

<button id = "start" style="background-color: #008000">Start</button>
<button id = "end" style="background-color: #800000">End</button>
<button id = "finish">Finish</button>
<script>

var canvas = document.querySelector('canvas')
var ctx = canvas.getContext('2d')

for (var x = 0.5; x < 301; x += 20) {
  ctx.moveTo(x, 0);
  ctx.lineTo(x, 300);
}

for (var y = 0.5; y < 301; y += 20) {
  ctx.moveTo(0, y);
  ctx.lineTo(300, y);
}

ctx.strokeStyle = "#000";
ctx.stroke();

var button = document.getElementById('finish')
var start_button = document.getElementById('start')
var end_button = document.getElementById('end')


start_button.onclick = ()=>{
    ctx.fillStyle="#008000";
  }

end_button.onclick = ()=>{
    ctx.fillStyle="#800000";
  }


var mouse = {x: 0, y: 0}
canvas.addEventListener('mousemove', function(e) {
  mouse.x = e.pageX - this.offsetLeft
  mouse.y = e.pageY - this.offsetTop
})

function getMousePos(canvas, evt) {
        var rect = canvas.getBoundingClientRect();
        return {
          x: evt.x - rect.left,
          y: evt.y - rect.top
        };
      }

function getNearestSquare(position) {
    var x = position.x;
    var y = position.y;

    if (x < 0 || y < 0) return null;
    x = (Math.floor(x / 20) * 20) + 0.5
    y = (Math.floor(y / 20) * 20) + 0.5
    return {x: x, y: y};
}

function containsObject(obj, list) {
    var i;
    for (i = 0; i < list.length; i++) {
        if (list[i].x === obj.x && list[i].y == obj.y) {
            return i;
        }
    }
    return -1;
}

borders = []

canvas.onmousedown = ()=>{
    var pos = getNearestSquare(getMousePos(canvas, mouse));
    if (pos != null) {
        index = containsObject(pos, borders)
        if (index != -1){
            borders.splice(index, 1)
            ctx.clearRect(pos.x + 1, pos.y, 19, 19)
        }
        else{
            ctx.fillRect(pos.x+1,pos.y+1,18,18);
            borders.push(pos)
        }
        ctx.fillStyle="#000000";
    }
}
var data = new Promise(resolve=>{
  button.onclick = ()=>{
    resolve(canvas.toDataURL('image/png'))
  }
})
</script>
"""

def draw(filename='drawing.png'):
  display(HTML(canvas_html))
  data = eval_js("data")
  print (data)
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return len(binary)
```

#### Σχολισαμός

To make it difficult for a hill-climbing algorithm to solve a maze, we need to construct a maze where there are multiple local maxima, and the optimal solution is far away from the initial starting point. In other words, the maze should have many possible paths that lead to local maxima that are not the global optimum.

One way to achieve this is to create a maze with many dead-ends and loops that look similar. If the hill-climbing algorithm chooses a path that leads to a dead-end or a loop, it will get stuck and won't be able to find a better solution. Another way is to add deceptive paths that lead to promising-looking but ultimately incorrect directions. This can confuse the algorithm and lead it to get stuck in a suboptimal solution.

#### Σχολιασμός:

Since Dijkstra does not take into account the distance from the goal, but only the actual distances from the starting point, it will start the search by examining all the nodes peripherally, a fact that will continue even after encountering the obstacle because it is not significantly affected by it's presence. On the other hand, the A\* algorithm takes into account both the actual distances from the starting point and the heuristic distances from the goal, resulting in a "diagonal" traversal that is interrupted by the presence of the obstacle. A\* bypasses the obstacle and continues its course to the right, but increasing its steps. However, A\* remains significantly faster than Dijkstra due to the targeted search it performs.

#### Σχολιασμός:

In general, the significant difference between A\* and Dijkstra's algorithm is the use of a heuristic function in A\* that guides the search towards the goal. When both A\* and Dijkstra's algorithm are applied to solve a maze with a unique path leading to the goal, they will follow the same number of steps and pass through the same nodes simultaneously. This happens because the distance of the nodes from the goal is not a significant piece of information, as it cannot change the search process, since there are no different neighbors to expand.
